#**1. Carga y Tratamiento de los Datos**

#**2. Encoding**

```python
#Se eliminan columnas que no se consideran relevantes
df = datos.drop(columns=['Partner','Dependents','PaperlessBilling','Charges.Total'])
df.sample(5)
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

columnas = X.columns
```

```python
columnas
```

```python
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),['MultipleLines',
                                'InternetService','OnlineSecurity','OnlineBackup',
                                'DeviceProtection','TechSupport','StreamingTV',
                                'StreamingMovies', 'Contract','PaymentMethod']),
                                  remainder='passthrough',
                                  sparse_threshold=0,
                                  force_int_remainder_cols=False)
```

```python
X = one_hot.fit_transform(X)
```

```python
one_hot.get_feature_names_out(columnas)
```

```python
pd.DataFrame(X, columns=one_hot.get_feature_names_out(columnas))
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

#**3. Modelando**

##*Separando datos*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=5)
```

##***3.1 Modelo Base***

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import recall_score

dummy = DummyClassifier()
dummy.fit(X_train, y_train)

#Sabiendo cuanto clasifico bien con los datos de prueba, precisión
#Sirve como punto de comparación
round(dummy.score(X_test, y_test),4)
```

##***3.2 Regresión Logística***

```python
#Llevando datos a una misma escala, 0 a 1
from sklearn.preprocessing import MinMaxScaler

#Normalizando X_train
normalizacion = MinMaxScaler()
X_train_norm = normalizacion.fit_transform(X_train)
X_test_norm = normalizacion.transform(X_test)
```

```python
y_previsto = model.predict(X_test_norm)
```

Para obtener una visualización más clara de la cantidad de features que son suficientes para el modelado, realizamos el siguiente análisis:

```python
def valores_metricas(y_test,y_pred_sel):
  exactitud = accuracy_score(y_test, y_pred_sel)
  precision= precision_score(y_test,y_pred_sel)
  recall = recall_score(y_test, y_pred_sel)
  f1 = f1_score(y_test, y_pred_sel)
  metricas = {
      'Exactitud':round(exactitud,4),
      'Precisión':round(precision,4),
      'Recall':round(recall,4),
      'F1':round(f1,4)
  }
  return metricas
```

```python
resultados = pd.DataFrame(index=['accuracy', 'precision', 'recall', 'f1'])
model_features = RandomForestClassifier(max_depth = 10, random_state=5, class_weight='balanced')
ct_features = [i if i !=0 else 1 for i in range(0,40,5)]

# Split the data into training and test sets before any encoding
X_train_df, X_test_df, y_train, y_test = train_test_split(df.drop('Churn', axis=1), y, stratify=y, random_state=5)

# Identify original categorical and numerical columns
categorical_cols = X_train_df.select_dtypes(include='object').columns
numerical_cols = X_train_df.select_dtypes(exclude='object').columns

for i in ct_features:
  selected_features = feature_importances['Feature'].values[:i]

  # Separate selected features into original categorical and numerical columns
  selected_categorical_cols = [col.split('__')[1] for col in selected_features if 'onehotencoder__' in col and col.split('__')[1] in categorical_cols]
  selected_numerical_cols = [col.split('__')[1] for col in selected_features if 'remainder__' in col and col.split('__')[1] in numerical_cols]
  # Include original categorical columns if they are in selected_features
  selected_categorical_cols.extend([col for col in selected_features if col in categorical_cols])
  # Include original numerical columns if they are in selected_features
  selected_numerical_cols.extend([col for col in selected_features if col in numerical_cols])


  # Apply one-hot encoding only to selected categorical columns
  one_hot_selected = make_column_transformer((OneHotEncoder(drop='if_binary'), selected_categorical_cols),
                                            remainder='passthrough',
                                            sparse_threshold=0,
                                            force_int_remainder_cols=False)

  X_train_encoded = one_hot_selected.fit_transform(X_train_df[selected_categorical_cols + selected_numerical_cols])
  X_test_encoded = one_hot_selected.transform(X_test_df[selected_categorical_cols + selected_numerical_cols])

  # Apply scaling
  X_train_scaled = normalizacion.fit_transform(X_train_encoded)
  X_test_scaled = normalizacion.transform(X_test_encoded)


  model_features.fit(X_train_scaled, y_train)
  y_pred_sel = model_features.predict(X_test_scaled)
  metricas = valores_metricas(y_test,y_pred_sel)
  resultados[i] = list(metricas.values())

resultados
```

**Se observa que un rango de 20 a 25 features es suficiente para que el modelo trabaje bien, pues a partir de estas, las métricas se mantienen constantes y ya no cambian significativamente. Esto nos muestra que no es necesario usar las 36 features (columnas de X), sino solo 20 a 25.**

Realizamos un balanceo de datos por *Undersampling* para verificar si favorece las métricas del modelo **Random Forest**. Es método intenta no eliminar información relevante para el modelo, por lo que se decide utilizarlo.

```python
from imblearn.under_sampling import NearMiss
from imblearn.pipeline import Pipeline as imbpipeline
```

```python
modelo_rf_3 = RandomForestClassifier(max_depth = 10, random_state=5)
pipeline = imbpipeline(steps=[('undersample', NearMiss(version = 3)), ('Random Forest', modelo_rf_3)])
skf = StratifiedKFold(n_splits=5, shuffle = True, random_state = 5)
cv_resultados_rf_2 = cross_validate(modelo_rf_3, X, y, cv=skf, scoring = ['accuracy', 'recall', 'precision', 'f1'])
cv_resultados_rf_2
```

##***5.1 Estrategias Propuestas***



Al analizar la importancia de las features en el modelo seleccionado (**Random Forest**) y el comportamiento de los boxplot en el análisis exploratorio, podemos percibir que las variables *Tenure* y *Cargos mensuales* son muy relevantes, pues los clientes que tienden a ser evasivos, tienen normalmente cargos mensuales altos, lo que lleva a que desde los primeros meses de contrato, deciden dejar el servicio.

Por otra parte, otras variables relacionadas al servicio propio podrían estar influenciando esta evasión, de forma general, se observa que las variables que contribuyen a tener menos servicios (*sin soporte técnico* o *seguridad online*, etc.) favorecen que los clientes se mantengan; y esto se relaciona directamente con cargos mensuales más bajos.

Con base en las conclusiones obtenidas, se proponen las siguientes estrategias:



1.   Favorecer el **Tipo de Contrato Mes a mes**, para que el cliente sienta la libertad de dejar la empresa cuando lo decida, esto fortalece la confianza en la empresa sin establecer plazos forzosos de contrato.
2.   Destacar el tipo de **Servicio de Internet por Fibra Óptica**, porque se percibe que es el modo de servicio que los clientes prefieren.
3.   Establecer paquetes de servicio más sencillos, en los que se ofrezca una menor cantidad de servicios (más no siempre es mejor), elegibles por los clientes, en los que el cargo mensual sea menor y favorezca la retención de clientes; esto analizando claramente que la empresa no se vea afectada.
